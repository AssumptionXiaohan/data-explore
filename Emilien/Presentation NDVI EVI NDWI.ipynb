```python
import gdal
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

NDVI Statistics:

```python
raster_path = 'NDVI_Summer.tif'
raster_dataset = gdal.OpenEx(raster_path, gdal.GA_ReadOnly)
geo_transform = raster_dataset.GetGeoTransform()
proj = raster_dataset.GetProjectionRef()

#Summer data
NDVIS = raster_dataset.GetRasterBand(1).ReadAsArray()
NDVISn0=NDVIS[NDVIS>0] #take out 0 for histograms + statistics
np.mean(NDVIS>0)
```

```python
raster_path = 'NDVI_Winter.tif'
raster_dataset = gdal.OpenEx(raster_path, gdal.GA_ReadOnly)
geo_transform = raster_dataset.GetGeoTransform()
proj = raster_dataset.GetProjectionRef()

#Winter data
NDVIW = raster_dataset.GetRasterBand(1).ReadAsArray()
NDVIWn0=NDVIW[NDVIW>0] #take out 0 for histograms + statistics
np.mean(NDVIW>0)
```

```python
f,ax = plt.subplots(nrows=2,ncols=1, figsize=(10,6))

n1, bins1, patches1=ax[0].hist(NDVISn0,bins=60,color='r')
ax[0].set_title('NDVI distribution (summer)')
n2,bins2,patches2=ax[1].hist(NDVIWn0,bins=60,color='b')
ax[1].set_title('NDVI distribution (winter)')

plt.show()
```

```python
#on same plot
f,ax = plt.subplots(figsize=(14,6))

ax.plot(bins1[:-1],n1,color='r')
ax.plot(bins2[:-1],n2,color='b')
```

A value of 0.4 seems to be a good value for a threshold low vegetation/high vegetation for NDVI to make a clear seperation between summer and winter: if an area has a big difference of NDVI between Summer and Winter, it is probably not a city. 

```python
NDVIS_thr=np.zeros([2826,2361],dtype='int')
NDVIS_thr[(NDVIS>0) & (NDVIS<0.4)]=1
NDVIS_thr[NDVIS>=0.4]=2

NDVIW_thr=np.zeros([2826,2361],dtype='int')
NDVIW_thr[(NDVIW>0) & (NDVIW<0.4)]=1
NDVIW_thr[NDVIW>=0.4]=2

f,ax = plt.subplots(nrows=1,ncols=2, figsize=(14,14))

ax[0].imshow(NDVIS_thr)
ax[1].imshow(NDVIW_thr)
plt.show()
```

```python
#Closer look at Bay area

ax[0].imshow(NDVIS_thr[1100:1400,400:600])
ax[1].imshow(NDVIW_thr[1100:1400,400:600])
f
```

We can also use NDVI=0.2 to compare, as seen in documentation from this website : https://www.agricolus.com/en/indici-vegetazione-ndvi-ndmi-istruzioni-luso/

```python
NDVIS_thr=np.zeros([2826,2361],dtype='int')
NDVIS_thr[(NDVIS>0) & (NDVIS<0.2)]=1
NDVIS_thr[NDVIS>=0.2]=2

NDVIW_thr=np.zeros([2826,2361],dtype='int')
NDVIW_thr[(NDVIW>0) & (NDVIW<0.2)]=1
NDVIW_thr[NDVIW>=0.2]=2

f,ax = plt.subplots(nrows=1,ncols=2, figsize=(14,14))

ax[0].imshow(NDVIS_thr)
ax[1].imshow(NDVIW_thr)
plt.show()
```

```python
#Closer look at Bay area

ax[0].imshow(NDVIS_thr[1100:1400,400:600])
ax[1].imshow(NDVIW_thr[1100:1400,400:600])
f
```
